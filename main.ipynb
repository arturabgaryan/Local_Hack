```python
for dot in range(78,50,-2):
    print(dot)
    for col in tqdm.tqdm(cases.columns):
        if col is not 'Date':
            mcases = Prophet()
            mdeaths = Prophet()
            mrecovery = Prophet()
            casesset = cases[['Date',col]].iloc[:(dot+1)].rename(columns={"Date":'ds', col:'y'})
            deathsset = deaths[['Date',col]].iloc[:(dot+1)].rename(columns={"Date":'ds', col:'y'})
            recoveryset = recovery[['Date',col]].iloc[:(dot+1)].rename(columns={"Date":'ds', col:'y'})
            mcases.fit(casesset)
            mdeaths.fit(deathsset)
            mrecovery.fit(recoveryset)
            future = mcases.make_future_dataframe(periods=7)
            forecastcases = mcases.predict(future.iloc[dot+7:dot+8])
            forecastdeaths = mdeaths.predict(future.iloc[dot+7:dot+8])
            forecastrecovery = mrecovery.predict(future.iloc[dot+7:dot+8])
            forecastcases.columns = [c+'_cases' for c in forecastcases.columns]
            forecastdeaths.columns = [c+'_deaths' for c in forecastdeaths.columns]
            forecastrecovery.columns = [c+'_recovery' for c in forecastrecovery.columns]
            merged = pd.concat([forecastcases, forecastdeaths,forecastrecovery], axis=1)
            datasetX = pd.concat([datasetX, merged])
            main_y.append([cases[col].iloc[dot+7], deaths[col].iloc[dot+7], recovery[col].iloc[dot+7]])
```
